```python
# data/raw/datos_clima/9170/datos_9170_2018-10-12T01:00:00UTC_2019-04-10T01:00:00UTC.json 
# data/raw/datos_clima/9170/datos_9170_2018-10-12_2019-04-10.json

# Vamos a renombrar los datos raw de nuestro repo DVC y Git
# Tenemos que modificar los .json de las subcarpetas de data/raw/datos_clima/ 
# Tenemos que quitar la parte de la hora que es la que da problemas

import os

# Definimos la ruta de la carpeta donde se encuentran los datos
ruta = "raw/datos_clima/"

# Recorremos las subcarpetas de la carpeta de datos
for subdirectorio in os.listdir(ruta):
    # Si el subdirectorio es una carpeta
    if os.path.isdir(ruta + subdirectorio):
        # Recorremos los archivos de la carpeta
        for archivo in os.listdir(ruta + subdirectorio):
            print(archivo)
            # Si el archivo es un .json o un .json.dvc
            if archivo.endswith(".json"):
                # Renombramos el archivo
                nuevo_nombre = archivo.replace("T01:00:00UTC_", "_").replace("T01:00:00UTC", "")
                os.rename(ruta + subdirectorio + "/" + archivo, ruta + subdirectorio + "/" + nuevo_nombre)
                print(f"Renombrando {archivo} a {nuevo_nombre}")
                
```
