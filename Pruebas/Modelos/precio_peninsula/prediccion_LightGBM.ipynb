```python
import pandas as pd
df = pd.read_parquet('data/processed/datos_precio/clima_precio_merged_recortado.parquet')
```

```python
df.head()
```

```python
# Eliminar todas las columnas que no contengan "temperatura" "viento" y "sol" en el nombre de la columna 

df_reducido = df.copy()
lista_particulas_mantener = [
    "tm", 
    "viento", 
    "sol", 
    "€/kwh", 
    "fecha", 
    "anio",
    "mes",
    "dia",
    "dia_semana",
    "estacion"]
for column in df.columns:
    if not any([particula in column for particula in lista_particulas_mantener]):
        df_reducido.drop(column, axis=1, inplace=True)
```

```python
df_reducido.head()
```

```python
# Get the unique years and find the last year
ultimo_anio = df_reducido['anio'].max()

# Create a new dataframe excluding the last year
df_entrenamiento = df_reducido[df_reducido['anio'] < ultimo_anio].copy()

print(f"Años incluidos en el conjunto de entrenamiento: {sorted(df_entrenamiento['anio'].unique())}")
print(f"Tamaño del conjunto de entrenamiento: {len(df_entrenamiento)} registros")

df_validacion = df_reducido[df_reducido['anio'] == ultimo_anio].copy()
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, r2_score
import mlflow
import mlflow.lightgbm
```

```python
# Preparar los datos
X = df_entrenamiento.drop(['€/kwh'], axis=1)

# Eliminar columnas datetime y object
datetime_object_columns = X.select_dtypes(include=['datetime64[ns]', 'object']).columns
X = X.drop(datetime_object_columns, axis=1)

# Asegurarse de que solo quedan columnas numéricas
X = X.select_dtypes(include=['int64', 'float64', 'bool'])

# Target variable
y = df_entrenamiento['€/kwh']

y_validacion = df_validacion['€/kwh']
```

```python

# Create lagged features for 2 days ahead prediction (48 hours as data is hourly)
def create_lagged_features(df, target_col, lag=48):
    df_copy = df.copy()
    df_copy[f'{target_col}_lag_{lag}'] = df_copy[target_col].shift(lag)
    return df_copy

# Prepare training data
df_train = create_lagged_features(df_entrenamiento, '€/kwh', lag=48)
df_train = df_train.dropna()  # Remove rows with NaN values

# Prepare validation data
df_val = create_lagged_features(df_validacion, '€/kwh', lag=48)
df_val = df_val.dropna()

# Guardar los datos de entrenamiento y validación
df_train.to_parquet('data/processed/datos_precio/df_train_lagged.parquet')
df_val.to_parquet('data/processed/datos_precio/df_val_lagged.parquet')
```

```python
from sklearn.preprocessing import StandardScaler

# Select features (excluding datetime, object columns and target)
feature_cols = df_train.select_dtypes(include=['int64', 'float64']).columns
feature_cols = feature_cols.drop('€/kwh')

# Prepare X and y for training
X_train = df_train[feature_cols]
y_train = df_train['€/kwh']

# Prepare X and y for validation
X_val = df_val[feature_cols]
y_val = df_val['€/kwh']

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
mlflow.set_experiment("precio/")

with mlflow.start_run(run_name="LightGBM_Model_Run"):
    # Configurar y entrenar el modelo
    params = {
        'boosting_type': 'gbdt',
        'objective': 'regression',
        'metric': 'rmse',
        'num_leaves': 31,
        'learning_rate': 0.1,
        'feature_fraction': 0.9,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'verbose': 0
    }
    
    # Ensure X_val has the same features as X
    X_val = X_val[X.columns]
    
    # Crear conjuntos de datos para LightGBM
    train_data = lgb.Dataset(X, label=y)
    
    # Entrenar modelo
    model = lgb.train(
        params,
        train_data,
        num_boost_round=100,
        valid_sets=[train_data]
    )
    
    # Hacer predicciones
    y_pred_train = model.predict(X)
    y_pred_val = model.predict(X_val)
    
    # Calcular métricas
    train_rmse = np.sqrt(mean_squared_error(y, y_pred_train))
    val_rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
    train_r2 = r2_score(y, y_pred_train)
    val_r2 = r2_score(y_val, y_pred_val)
    
    # Registrar parámetros
    for key, value in params.items():
        mlflow.log_param(key, value)
    mlflow.log_param('num_boost_round', 100)
    
    # Registrar métricas
    mlflow.log_metric('train_rmse', train_rmse)
    mlflow.log_metric('val_rmse', val_rmse)
    mlflow.log_metric('train_r2', train_r2)
    mlflow.log_metric('val_r2', val_r2)
    
    # Guardar el modelo
    input_example = X.iloc[:5]
    mlflow.lightgbm.log_model(model, "lightgbm_model", input_example=input_example)
    
    # Imprimir resultados
    print(f'Training RMSE: {train_rmse:.4f}')
    print(f'Validation RMSE: {val_rmse:.4f}')
    print(f'Training R2: {train_r2:.4f}')
    print(f'Validation R2: {val_r2:.4f}')
```
